5. Возьмем еще четыре гена и проверим, сколько мутаций у донора в них. Я выбрал следующие:

A) **KRAS** - этот ген также часто мутирует в раке и может играть роль в онкогенезе.

Б) **TCP11** - Несколько исследований показали, что ген TCP11 может быть связан с различными заболеваниями, включая онкологические заболевания, такие как рак легкого, рак молочной железы и рак предстательной железы, а также нейрологические заболевания, такие как биполярное расстройство и шизофрения.

В) **CDH13** - Исследования связывают ген CDH13 с различными типами рака. Например, сниженный уровень экспрессии гена CDH13 в связи с метастазированием рака молочной железы, а также обнаружено уменьшение экспрессии гена CDH13 у пациентов с раком легкого, раком толстой кишки и другими видами рака.

Одна из возможных причин связи между геном CDH13 и раком может быть связана с тем, что белок T-cadherin участвует в регуляции роста и миграции клеток, что является ключевыми процессами, связанными с развитием рака.

Г) **MBNL1** - это ген, который кодирует белок, играющий ключевую роль в сплайсинге РНК (RNA splicing) - процессе, который регулирует, как гены транскрибируются в РНК и как они переводятся в белки. Белок MBNL1 связывается с мотивами на мРНК и регулирует альтернативный сплайсинг, который может приводить к изменению функции и стабильности мРНК, а также к изменению свойств белков.
